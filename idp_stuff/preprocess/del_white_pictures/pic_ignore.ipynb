```
import os
cpt = sum([len(files) for r, d, files in os.walk("/content/drive/My Drive/resized_Images/bigDataSet/negative_1250")])
print(cpt)
```

```
#Define pictures directory
folders = glob.glob("/content/drive/My Drive/resized_Images/bigDataSet/negative_1250")
imagenames_list = []

#Get all images to a list
for folder in folders:
    for f in glob.glob(folder + '/*.png'):
        imagenames_list.append(f)
```

```
import pandas as pd
df = pd.DataFrame(imagenames_list)
print(df)
```

```
import cv2
import numpy as np


count = 0
for imgCounter in range(len(imagenames_list)):
#img =Image.open('/content/drive/My Drive/IDP/async/Capture.JPG')
  img = Image.open(imagenames_list[imgCounter])
  img1 = cv2.imread(imagenames_list[imgCounter], cv2.IMREAD_GRAYSCALE)
  n_white_pix = np.sum(img1 == 255)
  width, height = img.size
  dim = width*height
  prop = (n_white_pix/dim)*100
  if(prop > 90.00):
    print("image number- "+str(imgCounter) + ": " + "White percentage: ", "{0:.2f}".format(prop) + "%")
    if os.path.exists(imagenames_list[imgCounter]):
      print("Deleting File... ")
      os.remove(imagenames_list[imgCounter])
    else:
      print("The file does not exist")
    count += 1

print(count)
```

```
perc = (count/cpt)*100
print("Percentage of pictures having more than 90% white pixels: ", "{0:.2f}".format(perc) + "%")
```
