```
import sys
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')
from selenium import webdriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```
init_list = df["URL"].tolist()
scheme = 'http://'
url_list = [scheme + x for x in init_list]
start_urls = url_list
```

```
driver.set_window_size(1920, 1080) 
#start_urls = ["https://www.androidworld.it/"]
ext = 0
i = 0
imgCounter = 1265
```

```

for i in range(len(start_urls)):
  try:
    driver.get(start_urls[i]) 
    time.sleep(2)
    driver.save_screenshot('/content/drive/My Drive/idp/nonStartUp/' + str(imgCounter) + '.png')
  except TimeoutException:
    print('Website not found: ' + str(start_urls[i]))
  i += 1
  imgCounter += 1

driver.close()
```
